```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Salary_Data.csv')
data.tail()
```

```python
print("Shape of the dataset:", data.shape)
print("Any missing values?\n", data.isnull().sum())
```

```python
x = data.iloc[:, :-1]
y = data.iloc[:, -1].values.reshape(-1, 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print("Predicted Salaries:\n", y_pred)
print("Actual Salaries:\n", y_test)
```

```python
plt.scatter(x_train['YearsExperience'], y_train, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('SALARY VS EXPERIENCE (Training Set)')
plt.xlabel('Experience in Years')
plt.ylabel('Salary in Rupees')
plt.grid(True)
plt.show()
```

```python
plt.scatter(x_test, y_test, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')  # use train line for visual consistency
plt.title('SALARY VS EXPERIENCE (Testing Set)')
plt.xlabel('Experience in Years')
plt.ylabel('Salary in Rupees')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data = pd.read_csv('chicago_taxi_extended_dataset.csv')
data.head()
```

```python
print("Shape of the dataset:", data.shape)
print("Any missing values?\n", data.isnull().sum())
```

```python
x = data[['trip_miles']]
y = data[['fare']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

print("Predicted Fares:\n", y_pred)
print("Actual Fares:\n", y_test.values)
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('Fare vs Trip Miles (Training Set)')
plt.xlabel('Trip Miles')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
plt.scatter(x_test, y_test, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('Fare vs Trip Miles (Testing Set)')
plt.xlabel('Trip Miles')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data = pd.read_csv('body_weight_dataset.csv')
data.head()
```

```python
print("Shape of the dataset:", data.shape)
print("Any missing values?\n", data.isnull().sum())
```

```python
x = data[['Height']]
y = data[['Weight']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print("Predicted Weights:\n", y_pred)
print("Actual Weights:\n", y_test.values)
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('Weight vs Height (Training Set)')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.grid(True)
plt.show()
```

```python
plt.scatter(x_test, y_test, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('Weight vs Height (Training Set)')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.grid(True)
plt.show()
```
